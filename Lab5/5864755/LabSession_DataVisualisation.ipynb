# Matplotlib

For this excercise, we have written the following code to load the stock dataset built into plotly express.

```python
stocks = px.data.stocks()
stocks.head()
```

## Question 1: 

Select a stock and create a suitable plot for it. Make sure the plot is readable with relevant information, such as date, values.  

```python
# YOUR CODE HERE
x = stocks.date
y = stocks.GOOG

fig = plt.figure(figsize=(12, 9))
gs = fig.add_gridspec(nrows=2, ncols=2)
#
ax = fig.add_subplot(gs[:,:])
ax.plot(x,y)
ax.set_title('GOOGLE')
ax.set_xlabel('date')
ax.set_ylabel('value')
ticks = range(0, len(stocks),20)
ax.set_xticks(ticks)

plt.show()





```

## Question 2: 

You've already plot data from one stock. It is possible to plot multiples of them to support comparison.  
To highlight different lines, customise line styles, markers, colors and include a legend to the plot.

```python
# YOUR CODE HERE
d = stocks.date
g = stocks.GOOG
a = stocks.AAPL
am = stocks.AMZN
fb = stocks.FB
nf = stocks.NFLX
m = stocks.MSFT

fig, ax = plt.subplots()


ax.plot(d,g, label = 'Google')
ax.plot(d,a, label = 'Apple')
ax.plot(d,am, label = 'Amazon')
ax.plot(d,fb, label = 'Facebook')
ax.plot(d,nf, label = 'Netfix')
ax.plot(d,m, label = 'Microsoft')

ax.set_title('stock')
ax.set_xlabel('date')
ax.set_ylabel('value')
ticks = range(0, len(stocks),20)
ax.set_xticks(ticks)
ax.legend()

plt.show()

```

## Question 5:

Recreate the barplot below that shows the population of different continents for the year 2007. 

*Hints:*

- Extract the 2007 year data from the dataframe. You have to process the data accordingly
- use [plotly bar](https://plotly.com/python-api-reference/generated/plotly.express.bar)
- Add different colors for different continents
- Sort the order of the continent for the visualisation. Use [axis layout setting](https://plotly.com/python/reference/layout/xaxis/)
- Add text to each bar that represents the population 

```python
#load data
df = px.data.gapminder()
df.head()
```

```python
# YOUR CODE HERE


df_2007 = df.query('year==2007')
df_2007_new = df_2007.groupby('continent').sum()
fig = px.bar(df_2007_new, x="pop", y=df_2007_new.index, color=df_2007_new.index, orientation='h', text_auto=True)
fig.update_yaxes(categoryorder='total descending')


fig.show()

```
